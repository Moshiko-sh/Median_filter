```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2

def median_filter(image, kernel_size):
    # Getting the dimensions of the image
    height, width = image.shape

    # Defining the output image
    filtered_image = np.zeros((height, width), dtype=np.uint8)

    # Calculating the padding size to handle the borders
    pad = kernel_size // 2

    # Applying the median filter
    for i in range(pad, height - pad):
        for j in range(pad, width - pad):
            # Extracting the region of interest
            region = image[i - pad:i + pad + 1, j - pad:j + pad + 1]

            # Applying the median operation
            median_value = np.median(region)

            # Setting the pixel value in the output image
            filtered_image[i, j] = median_value

    return filtered_image

# Reading the image
image = cv2.imread('/content/before.png', cv2.IMREAD_GRAYSCALE)

# Applying median filter with kernel size 5 (for example)
filtered_image = median_filter(image, kernel_size=5)

# Displaying the original image
print("Original Image:")
cv2_imshow(image)
print("")
# Displaying the filtered image
print("Filtered Image:")
cv2_imshow(filtered_image)
```
